```python
#Drop the duplicates
df=df.drop_duplicates()
```

```python
df
```

```python
#Remove Unwanted column
df=df.drop(columns ="Not_Useful_Column") # to save assign to variable or inplace=True
```

```python
df
```

```python
#Remove unwanted characters using strip
df["Last_Name"]=df["Last_Name"].str.strip("...,/,_")    #we can also use lstrip or rstrip
```

```python
df
```

```python
df
```

```python
#df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
```

```python
#Correct the format of Phone_number by removing non_numeric characters: Using replace
df["Phone_Number"] = df["Phone_Number"].str.replace(r'[^a-zA-Z0-9]', '', regex=True)
```

```python
df
```

```python
#Format ThE Phone num col
for number in df["Phone_Number"]:
    number = str(number)  # convert it to string since the col has NaN
    print(number[0:3] + '-'+ number[3:6] + '-' + number[6:])

df
```

```python
# Or we can use Lambda function:
#df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x)[0:3] + '-' + str(x)[3:6] + '-' + str(x)[6:10])
```

```python
# Remove Na--/ nan--
df["Phone_Number"]=df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"]=df["Phone_Number"].str.replace('Na--','')

df
```

```python
df[["Street","State","Zip_code"]]=df["Address"].str.split(",",n=2,expand = True) #got error because didn't write n=
                                                                                 #got syntax error: df[[ double brackets ]]
df
```

```python
#Replacing Y/N with Yes/No respectively
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')


df["Paying Customer"] = df["Paying Customer"].str.replace('Y','Yes')
df["Paying Customer"] = df["Paying Customer"].str.replace('N','No')
df["Paying Customer"] = df["Paying Customer"].str.replace('No/a','N/a')

df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')


df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Y','Yes')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('N','No')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No/a','N/a')

df
```

```python
#Fill N/a and NaN(technically they are blank) with empty string
df= df.fillna('')
df
```

```python
#Remove all the rows with Yes in Do_Not_Contact col
for x in df.index:     # x --> Index Number
    if df.loc[x,"Do_Not_Contact"]=='Yes':   #df.locotion[index_num,"col_name"]
         df.drop(x,inplace=True)
   
```

```python
#Remove all the rows with null value in phone number
for x in df.index:     # x --> Index Number
    if df.loc[x,"Phone_Number"]=='':
         df.drop(x,inplace=True)

#another way to remove null values
#df=df.dropna(subset="Phone_Number")    #na -->null
df
```

```python
#Reset the index
df=df.reset_index(drop=True)  #drop = True to remove original index
```
